```python

"""
Docstring:

Introduction: This game is based off the Disney Pixar movie 'Finding Nemo'. 
Before the events of the game, you were once a married clownfish with hundreds
of fresh eggs and soon-to-be baby clownfish, until a baraccuda ate your wife
and all but one egg, your now only son, Nemo.

Round 1: Decide to follow Dory or not.
Round 2: Trust the sharks or not.
Round 3: Deal with the trench.
Round 4 (Optional): Find out how to pass the jellyfish.
Round 5: Enter the correct address found on the goggles.

Known bugs/errors:
1: Some text needs formatted properly
2: Randomly given hints will sometimes repeat

"""

import time
import random

def start_room1():
    
    print("""
    Your son Nemo has just been scooped up by scuba divers after his
friends dared him to 'touch the butt' over the dropoff. In your
panicked state you immediately swim out into the dropoff to chase
after the boat that now has your son, but the boat speeds off much 
faster than you are able to swim and dissappears above the water. 
Seeing the boat escape your view puts you into a manic state of 
worry and you begin screaming if anyone has seen a boat.

          
    One fish, a pacific blue tang named Dory, rams into you head first and,
after asking if you are okay, says that she has seen a boat recently and 
that you should follow her.\n""")
    # Explains the intro to the story to the player and gives the first hurdle
    
    time.sleep(2)
    # Waits 2 seconds before showing the input and print boxes
    
    input(prompt = '<Press enter to view your options>\n')
    # Tell player to hit enter to view options, gives player time to read the story
    
    print("Enter 1 to follow Dory.")
    print("Enter 2 to ignore Dory and keep searching on your own.")
    print("Enter 3 to return back to the anemone and do nothing.\n")
    # Lists the options that the player can choose from
    
    choice = input(prompt = "<Enter an option between 1-3>\n")
    # Tell player that the acceptable options are between 1-3
    
    choice = choice.lower()
    # Convert players input into lowercase
    
    # Checks if player entered '1' or 'follow' in their input
    if '1' in choice or 'follow' in choice:
        print(
    "You decide to follow Dory and stick closely behind her.")
        time.sleep(2)
        # Waits 2 seconds before showing print boxes and moving the player into the next room
        print("""
    You decide to follow Dory because, at the moment, she's the
only lead you have. After many sporadic and seemingly random twists and turns, 
Dory stops, turns around, and begins to berate you for following her. 
Apparently she thinks that you've been stalking her, but after some discussion 
it is revealed that she has short term memory loss and you agree that she can 
still help you find your son.""")
        dory_room2() 
        # Move player into the next dory_room2()

        # Checks if player entered '2' or 'ignore' or 'keep searching' in their input
    elif '2' in choice or 'ignore' in choice or 'keep searching' in choice:
        print("""
    You ignore Dory's request to follow her and you swim in the
other direction, never finding your son.""")
        fail() 
        # Move player into the fail() room
        
        # Checks if player entered '3' or 'return' or 'nothing' in their input
    elif '3' in choice or 'return' in choice or 'nothing' in choice:
        print("""
    The heartbreak of having your son missing is too much to bear
so you return back to your anemone to sulk, but a feeling of guilt 
comes over you and you decide to start the search again.""")
        start_room1() 
        # Move player back to start_room() for them to reselect an option
        
        # If player input doesn't have any of the options have them retry the room
    else:
        print("Invalid entry. Please try again.\n")
        start_room1() 
        # Move player back to the start_room() for them to reselect an option that is valid
        
    
def dory_room2():
    
    print("""
    While swimming around searching for Nemo you come across three sharks that
claim 'fish are friends, not food' and invite you to one of their meetings in 
a submarine.\n""") 
    # Explains the next part of the story to the player and gives the next hurdle

    input(prompt = '<Press enter to view your options>\n')
    # Tell player to hit enter to view options, gives player time to read the story
    
    print("Enter 1 to ignore the sharks and continue searching on your own.")
    print("""Enter 2 to tell the sharks that you don't believe them and that
        their fathers would be dissapointed with their lying.""")
    print("Enter 3 to go with the sharks to their meeting.\n")
    
    choice = input(prompt = "<Enter an option between 1-3>\n ")
    # Tell player that the acceptable options are between 1-3
        
    choice = choice.lower()
    # Convert players input into lowercase
    
    # Checks if player entered '1' or 'ignore' or 'continue' or 'searching' in their input
    if '1' in choice or 'ignore' in choice or 'continue' \
    in choice or 'searching' in choice:
        print("""
    You ignore the sharks and continue searching aimlessly, maybe 
the sharks could have helped?\n""")
        dory_room2() # Have player restart the room

    # Checks if player entered '2' or 'tell' or 'believe' in their input    
    elif '2'in choice or 'tell' in choice or 'believe' in choice:
        print("""
    It was a bad move insulting the sharks. The largest one, a 
great white named Bruce, never knew his father, so he enters a depressed
manic state and eats you and Dory.\n""")
        fail() # Move the player into the fail room
    
    # Checks if player entered '3' or 'go' or 'meeting' in their input
    elif '3' in choice or 'go' in choice or 'meeting' in choice:
        print("""
    You feel that the sharks are trustworthy so you decide to go 
with them to the meeting in the sunken submarine.\n""")
        time.sleep(2)
        # Waits 2 seconds before showing the input and print boxes
        print("""
    You attend the meeting with the sharks in the sunken submarine but remain
very cautious, they are sharks after all, when in the corner of your eye you
spot a divers mask that fell off of the boat that took Nemo. There's writing
on it but you don't know how to read, fortunately enough though, Dory does! 
You show it to her and she tells you that it reads 'P Sherman, 42 Wallaby Way,
Sydney'. That could be important to remember. 
    Soon after, Dory experiences a cut that leaks blood which triggers the 
sharks to attack. You and Dory manage to detonate an old torpedo in the 
submarine which sinks the mask, but you and Dory are able to make an escape.
""")
        shark_room3() # Move the player into room 3
     
    # If player input doesn't have any of the options have them retry the room
    else:
        print("Invalid entry. Please try again.\n")
        input(prompt = '<Press enter to retry>\n')
        dory_room2() # Have player restart the room to enter a valid input



def shark_room3():

    print("""
    After your escape you come across a school of moonfish who advise you to 
swim through the scary trench just ahead, not over it.\n""")
     # Explains the next part of the story to the player and gives the next hurdle
    
    
    input(prompt = '<Press enter to view your options>\n')
    # Tell player to hit enter to view options, gives player time to read the story
    
    print("Enter 1 to listen to the moonfish and swim through the trench.")
    print("Enter 2 to disregard the moonfish and swim over the trench.")
    print("Enter 3 to try to find a different route that avoids the trench.\n")
    # Lists the options that the player can choose from
    
    choice = input(prompt = "<Enter an option between 1-3>\n")
    # Tell player that the acceptable options are between 1-3
        
    choice = choice.lower()
    # Convert players input into lowercase
    
    # Checks if player entered '1' or 'listen' or 'through' in their input
    if '1' in choice or 'listen' in choice or 'through' in choice:
        print("""
    You smartly listen to the moonfish and safely swim through
the scary trench.\n""")
        time.sleep(2)
        # Waits 2 seconds before showing print boxes and moving the user to the next room
        print("""
    You successfully make it past the trench and are nearly to your
location. Word has gotten out that you have been searching for your son and 
will do anything to get him back, including blowing up some sharks! Half of 
ocean is now on your side which you use to your advantage. You hitch a ride on
some totally rad turtles in the East Australian Current that take you all the 
way to the Sydney Harbour.""")
        win_room4()
        # Move the player to the last room, the win_room() 
    
    # Checks if player entered '2' or 'disregard' or 'over' in their input
    elif '2' in choice or 'disregard' in choice or 'over' in choice:
        print("""
    You ignore the moonfish and decide to swim over the trench.
Everything seems fine at first, until you find yourself before a bloom of 
jellyfish that you must now deal with.\n""")
        jelly_room()
        # Move the player to the jelly_room()
    
    # Checks if player entered '3' or 'different' or 'avoid' in their input
    elif '3' in choice or 'different' in choice or 'avoid' in choice:
        print("""
    You and Dory look around for another route and fail to find
one. It looks like you must deal with the trench.\n""")
        shark_room3()
        # Have the player replay the current room.
    
    # If player input doesn't have any of the options have them retry the room
    else:
        print("Invalid entry. Please try again.\n")
        input(prompt = '<Press enter to retry>\n')
        dory_room2() # Have player restart the room to enter a valid input

        
def jelly_room():
# Player only enters this room if they decide to swim over the trench

    # Randomly generate the number of guesses the player has
    guesses = random.randint(2, 7)
    
    # A list of messages that could be printed based on the users input
    jelly_fail_msg = ["""
    You try to slowly dodge the jellyfish tentacles but get stung and quickly
back out.\n""",
"""You attempt to swim under the jellyfish but they extend down into the
trench so you back out.\n""",
"""You try to speak to the jellyfish but they don't respond, maybe because
they don't have mouths.\n""",
"What kind of person are you? You can't sacrifice Dory!\n",
"""You try to swim over the jellyfish but they extend all the way to the 
surface.\n""",
'The jellyfish are unphased by any of your actions and stay put.',
'... You cannot be serious...\n',
"""You try to wait for the jellyfish to move, but after waiting it's 
clear that they won't be moving any time soon.\n"""]
    
    # A list of potential actions that will be randomly pulled from
    jelly_actions = ['to try to slowly swim through the jellyfish.',
                     'to try to swim under the jellyfish.',
                     'to try to speak to the jellyfish and tell them to move.',
                   """to use Dory as a sacrificial meat shield and power
                      through the jellyfish.""",
                     'to try to swim over the jellyfish.',
                     'to try to distract the jellyfish and swim past them.',
                     'to join the jellyfish colony and start a new life.']
   
    # This while function will run until guesses = 0, it doesn't matter what
    # answer you give, but after 3 guesses it will put you into the next room
    while guesses > 0:
        dummy_action = random.choice(jelly_actions)
        input(prompt = "<Press enter to view your options>\n")
        # Tell player to hit enter to view options, gives player time to read the story
        
        print(f"Enter 1 {dummy_action}")
        print("Enter 2 to wait for the jellyfish to move elsewhere.\n")
        
        choice = input(prompt = "<Enter an option between 1-2>\n")
        # Tell player that the acceptable options are between 1-2
        
        choice = choice.lower()
        # Convert players input into lowercase
        
        # No matter what choice it will print a temporary fail message and deduct 1 from the guess total
        # Additionally, checks for what action was generated and displays the appropriate fail message
        if dummy_action == jelly_actions[0]:
            if '1' in choice or 'through' in choice:
                print(jelly_fail_msg[0])
                guesses -= 1
            elif '2' in choice or 'wait' in choice:
                print(jelly_fail_msg[7])
                guesses -= 1
        elif dummy_action == jelly_actions[1]:
            if '1' in choice or 'under' in choice:
                print(jelly_fail_msg[1])
                guesses -= 1
            elif '2' in choice or 'wait' in choice:
                print(jelly_fail_msg[7])
                guesses -= 1
        elif dummy_action == jelly_actions[2]:
            if '1' in choice or 'speak' in choice or 'move' in choice:
                print(jelly_fail_msg[2])
                guesses -= 1
            elif '2' in choice or 'wait' in choice:
                print(jelly_fail_msg[7])
                guesses -= 1
        elif dummy_action == jelly_actions[3]:
            if '1' in choice or 'meat' in choice or 'sacrifice' in choice \
            or 'power' in choice or 'shield' in choice:
                print(jelly_fail_msg[3])
                guesses -= 1
            elif '2' in choice or 'wait' in choice:
                print(jelly_fail_msg[7])
                guesses -= 1
        elif dummy_action == jelly_actions[4]:
            if '1' in choice or 'over' in choice:
                print(jelly_fail_msg[4])
                guesses -= 1
            elif '2' in choice or 'wait' in choice:
                print(jelly_fail_msg[7])
                guesses -= 1
        elif dummy_action == jelly_actions[5]:
            if '1' in choice or 'distract' in choice or 'past' in choice:
                print(jelly_fail_msg[5])
                guesses -= 1
            elif '2' in choice or 'wait' in choice:
                print(jelly_fail_msg[7])
                guesses -= 1
        elif dummy_action == jelly_actions[6]:
            if '1' in choice or 'join' in choice or 'new life' in choice:
                print(jelly_fail_msg[6])
                guesses -= 1
            elif '2' in choice or 'wait' in choice:
                print(jelly_fail_msg[7])
                guesses -= 1
        else:
            print("Invalid entry. Please try again.\n")
            input(prompt = '<Press enter to retry>\n')
            jelly_room() # Have player restart the room to enter a valid input
        
    time.sleep(3)
    # Waits 3 seconds before showing the print
    
    print("""
    You sit back from the wall of jellyfish and try to think of a way to get
around...""")
    
    time.sleep(3)
    # Waits 3 seconds before showing the print
    
    print("""
    It finally dawns over you that you can bounce on the tops of the jellyfish
and they won't sting you! Knowing that Dory is gullible you challenge her to a
race across the tops of the jellyfish to the other side of the bloom.\n""")
    
    time.sleep(3)
    # Waits 3 seconds before showing the print and moving you to the next room
    
    print("""
    You successfully make it past the trench and are nearly to your
location. Word has gotten out that you have been searching for your son and 
will do anything to get him back, including blowing up some sharks! Half of 
the ocean is now on your side which you use to your advantage. You hitch a
ride on some totally rad turtles in the East Australian Current that take you
all the way to the Sydney Harbour.""")
    
    win_room4()
    
    
def win_room4():

    print("""
    At the harbour you meet a pelican that has offered to give you a lift to 
your final destination if you give them the address. You do remember it right?
\n""")
    # Explains the next part of the story to the player and gives the next hurdle
    
    # Correct answer we are looking for
    address = '42 WALLABY WAY'
    # Number of guesses the player has
    guesses = 3
    
    # A list of hints that will be displayed if the player enters the wrong address
    hint_list = ["The street number is the meaning of life.",
               """The street name is an animal that looks like a small 
                  kangaroo.""",
                 "The type of roadWAY is a tricky one.",
                 "The street number was Jackie Robinson's number.",
               """The street name shares a name with what you'd call a member
                  of the Australia National Rugby team."""]
    
    # The while loop will run until the player runs out of guesses or if they
    # enter the address correctly
    while guesses > 0:
    
        input(prompt = "<Press enter to continue.>\n")
        # Tell player to hit enter to view options, gives player time to read the story
    
    # Ask for players input in the form of a street name and number
        user_address = input(prompt = """
<Fill in the street number and name.>
    ex: 123 Main Street \n""")
        
        # If the player guesses wrong subtract 1 from the guess total
        if user_address.upper() != address:
            guesses -= 1
        # Tell player they guessed wrong and tell them how many guesses are left
            print(f"""
    You entered the wrong address! You have {guesses} guesses remaining.\n""")
        # Randomly give the player a hint from the hint_list 
            if guesses > 0:
                print(f"""
    Here is a hint!\n
    {random.choice(hint_list)}""")
            # If player runs out of guesses they fail
            elif guesses == 0:
                print("""
    You were wrong too many times so the pelican gets fed up and puts you back
into the harbour. You came so far yet still failed to find Nemo.\n""")
                fail()
                break
        # Ends the while loop if the user enters the incorrect address too many
        # times and send them to the fail() room
    
        # If the player guesses correctly tell them and finish the story
        elif user_address.upper() == address:    
            print("You are correct!\n")
            print("""
    The friendly pelican flies you to the window of the location, a dentist
office, and you find Nemo in the fish tank with his new fish buddies. You wait
until the dentist leaves the room and then instruct the pelican to scoop up
Nemo and his friends and get back to the ocean.\n""")
            print("""
    Congratulations on reuniting with your son and beating the game!""")  
            break
        # Ends the while loop if the user enters the correct address
                  
    

def fail():
    print("You've failed to find your son Nemo, but thanks for playing!")
    input(prompt = '<Press any key to exit>\n')
    # Display when the player fails the game
    


start_room1() # EVERYTHING goes before this clause
```
